### Top 10 Genre by Popularity

```python
plt.title('Top 10 Genre by Popularity')
t=spotify.sort_values('Popularity',ascending=False)
top=t.head(10)
sns.barplot(data=top,x='Popularity',y='Genre')
plt.grid(c='k')
plt.show()
```

### Loudness vs Energy

```python
sns.boxenplot(data=spotify,y='Energy',x='Loudness..dB..')
plt.grid(ls=':',c='k')
plt.title('Loudness vs Energy')
plt.show()
```

### Top 10 Track name by Popularity

```python
print('Top 10 Track name by Popularity')
t1=spotify.sort_values('Popularity',ascending=False)
top1=t.head(10)
sns.jointplot(data=top1,x='Popularity',y='Artist.Name',color='g')
plt.grid()
plt.show()
```

### Top 5 Track name by Beats.Per.Minute

### Beats.Per.Minute vs Genre

```python
plt.scatter(spotify['Beats.Per.Minute'],spotify['Genre'])
plt.xlabel('Beats.Per.Minute')
plt.ylabel('Genre')
plt.title('Beats.Per.Minute vs Genre')
plt.grid(ls=':',c='k')
plt.show()
```

### Liveness vs Speechiness

```python
plt.scatter(spotify['Liveness'],spotify['Speechiness.'],color='r')
plt.xlabel('Liveness')
plt.ylabel('Speechiness')
plt.title('Liveness vs Speechiness')
plt.grid(c='k')
plt.show()
```

### Acousticness & Popularity show in Density

```python
spotify['Acousticness..'].plot(kind='kde',label='Acousticness')
spotify['Popularity'].plot(kind='kde',label='Popularity')
plt.legend()
plt.grid()
plt.title('Acousticness & Popularity show in Density ')
plt.show()
```
