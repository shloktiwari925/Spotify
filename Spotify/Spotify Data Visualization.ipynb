```python
plt.figure(figsize=(12,6))
sns.heatmap(sp.corr(), annot=True)
plt.show()
```

```python
sp.head()
```

```python
value = sp['Genre'].value_counts()
```

```python
sp1 = sp[sp['Genre'].isin(value.index[value > 1])]
sp1.shape
```

```python
sp1.info()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Artist.Name', y='Beats.Per.Minute', data = sp1)
plt.xticks(rotation=90)
plt.title('Singer with their Rate of Beat Drop')
plt.grid(ls=':', c='y')
plt.xlabel('Name of Singers')
plt.ylabel('Beat Drop Rate')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Genre', y='Beats.Per.Minute', data = sp1)
plt.xticks(rotation=90)
plt.title('Genre with Rate of Beat Drop')
plt.grid(ls=':', c='k')
plt.xlabel('Name of Genre')
plt.ylabel('Beat Drop Rate')
plt.show()
```

```python
plt.figure(figsize=(13,7))
sp1['Genre'].value_counts().plot(kind='pie',autopct='%0.2f%%')
plt.title('Genre count')
plt.ylabel('Types of Genre')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='Artist.Name',y='Energy',data=sp1)
plt.xticks(rotation=90)
plt.title('Energy level of Singers')
plt.xlabel('Singers')
plt.grid(ls=':',c='k')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='Artist.Name', y='Danceability', data=sp1, hue='Genre')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation = 90)
plt.title('Danceability of songs according to singers and genre')
plt.xlabel('Singers')
plt.grid(ls=':', c='y')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x='Artist.Name', y='Popularity', data=sp1, hue='Genre')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation = 90)
plt.title('Popularity of singers')
plt.xlabel('Singers')
plt.grid(ls=':', c='y')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x = 'Liveness', y = 'Speechiness.' ,data =sp1,c='r', hue='Genre')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.xlabel('Liveness')
plt.ylabel('Speechiness')
plt.title('Liveness vs Speechiness')
plt.grid(ls=':',c='k')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(x= 'Speechiness.' , y='Beats.Per.Minute', data=sp1, c='r')
plt.title('Speechiness according to beat drop')
plt.grid(ls=':', c='k')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x = 'Energy', y = 'Loudness..dB..' ,data =sp1,c='y')
plt.xlabel('Energy')
plt.ylabel('Loudness..dB..')
plt.title('Loudness..dB.. vs Energy')
plt.grid(ls=':',c='k')
plt.show()
```
